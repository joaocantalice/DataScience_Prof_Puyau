# O que é Data Science?

__Definição de mercado__: é o processo de exploração, manipulação e análise dos dados para a __descoberta__ e __previsão__ através da criação de _hipóteses_, _testes_ e _validação_ com o objetivo de responder e/ou fazer recomendações capazes de serem diferenciais de negócio.

Todo esse processo precisa ter um forte __embasamento estatístico__ e __matemático__ e ser __diretamente ligado ao negócio__, além de poder ser feito de forma __escalável__ e __replicável__.

O atributo `columns` retorna uma lista com o nome de todas as colunas do _dataframe_.

### Pegando as colunas

```python
df.columns
```

Podemos melhorar o tipo de dado das colunas, para otimizar o uso de memória. Isto é muito útil, quando se está trabalhando com um conjunto de dados muito grande. 

__OBS__: durante o tratamento e o cálculo de dados, o dataframe fica em memória

Neste caso, a coluna `preço da garrafa` está como `float64`, o que ocupa muito espaço em memória.

A biblioteca __PANDAS__ nos ajuda neste sentido e fornece o método `to_numeric()`. Com este método, você passa a coluna e o parâmetro `downcast` que fará a convesão, sempre que possível, para o tipo de dado especificado. 

__Dica__:

* use `integer` para colunas com valores numéricos inteiros
* use `float` para colunas com valores numéricos com ponto flutuante

### Alterando o tipo de dado de uma coluna numérica

```python
df['preço da garrafa'] = pd.to_numeric(df['preço da garrafa'], downcast='float')
```

Após o uso do método `to_numeric`, chamamos novamente o método `info` e percebemos que, agora, a coluna `preço da garrafa` está como `float32`, um tipo de dados mais adequado para os valores desta coluna. 

__OBS__: com isso diminuímos o uso de memória do _dataframe_ em questão.

```python
df.info()
```

Agora que aprendemos a usar o método `to_numeric`, vamos aplicar, também, para a coluna `quantidade de garrafas`. 

Caso não se lembre, exatamente do nome da coluna, ou seja, como está escrita, recomenda-se invocar o atributo `columns` e a partir do retorno deste atributo, copiar o nome da coluna que desejamos. 

```python
df.columns
```

Aqui usamos o método `to_numeric` na coluna `quantidade de garrafas`, fazendo o _downcast_ para `integer`. 

```python
df['quantidade de garrafas'] = pd.to_numeric(df['quantidade de garrafas'], downcast='integer')
```

Repare que agora, a coluna `quantidade de garrafas` está como `int8` ao invés do `int64` quando carregamos o _dataframe_ em memória. 

```python
df.info()
```

### Método Describe

Para conhecermos ainda mais o _dataset_ em que estamos trabalhando, podemos utilizar a função `describe`. 

O método describe() que nos dá, inicialmente, um resumo sobre as **Series** numéricas com as seguintes informações:

* __count__ : mostra quantos registros / linhas há em nosso dataframe
* __mean__ : mostra a média. Vale lembrar que a depender da coluna, esse valor pode não fazer muito sentido mas no nosso exemplo, a média se aplica as colunas de quantidade de garrafas, preço da garrafa e total.
* __std__ : desvio padrão, esse tópico é abordado na especialização em Data Science
* __min__ : valor mínimo
* __os quartis__: 
    * __25%__ : 1º quadrante, significa que 25% dos dados estão abaixo de determinada quantidade. Ex: Olhando para a coluna de quantidade de garrafas, temos o valor de 5. Isso significa dizer que 25% dos registros do nosso dataframe estão abaixo de 5 garrafas.
    * __50%__ : 2º quadrante, também conhecido como mediana. Metade dos registros do nosso dataframe estão abaixo de 8 garrafas, ainda olhando para a coluna de quantidade de garrafas.
    * __75%__ : 3º quadrante, onde temos 75% dos registros do nosso dataframe abaixo de 10 garrafas.
* __max__ : valor máximo

```python
df.describe()
```

Pode-se utilizar este método em uma **Series** numérica específica e, também, numa **Series** de strings, onde as informações retornadas são diferentes das **Series** numéricas. Assim, obtemos os seguintes valores:

* __count__ : retorna quantas linhas que há no nosso dataframe
* __unique__ : retorna os valores únicos encontrados nesta Serie específica
* __top__ : retorno o "termo" encontrado com maior frequência. Vale ressaltar que esta informação retornará a primeira ocorrência, caso haja, mais termos com a mesma frequência.
* __freq__ : retorna o número de ocorrências do no termo listados em __top__

```python
df['estilo'].describe()
```
